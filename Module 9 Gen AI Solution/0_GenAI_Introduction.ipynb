### What is Generative AI?

Generative AI is the branch of artificial intelligence that doesn’t just analyze existing data—it creates new content from it. Instead of predicting a number or classifying an input, generative models learn the underlying patterns of language, images, code, or audio and then produce something completely new: a paragraph, a SQL query, a poem, a design, a synthetic dataset, a marketing copy, a personalized recommendation, or even executable logic for an agent.

At its core, Generative AI is the shift from AI that answers to AI that generates.

It learns from billions of examples and builds a deep understanding of structure, relationships, and meaning. When prompted, it uses that understanding to generate coherent, context-aware, and human-like outputs. This makes it a breakthrough for students—because for the first time, they can talk to a computer and the computer can create with them.

### What is Gen AI Use Cases?

Generative AI use cases are the practical, real-world ways we apply models like DBRX, Llama 3, or GPT-style transformers to create value. The power of GenAI lies in its ability to generate, reason, summarize, extract, classify, and increasingly take actions—not just converse.

For students learning on Databricks Free Edition, think of use cases in four broad layers that mirror how enterprises adopt GenAI:

### 1. Knowledge & Text Understanding

The model reads, understands, and rewrites information.
This is the foundation.

Examples include:

- Summarizing long documents
- Explaining complex concepts in simple words
- Extracting structured fields from unstructured text
- Translating text between languages
- Converting messy raw text into clean datasets

On Databricks, this becomes powerful when connected with PDF ingestion, Delta tables, and AI Functions like ai_extract.

### 2. Content Generation

The model creates new content based on instructions.

Examples:

- Writing blogs, product descriptions, emails, or marketing copy
- Drafting SQL queries or PySpark code
- Auto-generating documentation or insights
- Creating synthetic datasets for ML training
- Students can do all of this with a free DBRX or Llama model via the Databricks Model Catalog.

### 3. Intelligent Decisioning & Automation

GenAI does reasoning—not just writing text.

Examples:

- Generating step-by-step recommendations
- Classifying customer sentiment or ticket priority
- Creating next-best-action workflows
- Detecting anomalies or patterns in logs
- Drafting business insights from Delta tables

With Databricks, these tasks become seamless using SQL + AI functions or Python notebooks.

### 4. Agentic Applications (the next big shift)

GenAI that does more than chat—it acts.

Examples:

- Agents that read policy PDFs and answer questions
- SQL agents that query Delta tables
- Retail insights agents that analyze sales data
- Multi-agent systems where one agent retrieves data and another reasons
- Genie apps that act as conversational BI tools

This is the direction the world is moving toward: from chatbots to autonomous systems that think, reason, and take actions.

In Summary, GenAI use cases include:

- Understanding text
- Generating content
- Automating decisions
- Building acting, reasoning agents tied to data

For students, these use cases define the entire journey ahead: how you will build prompts, apps, and agents in Databricks Free Edition—connecting the power of GenAI with the power of the Lakehouse.
