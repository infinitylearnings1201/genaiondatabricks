````python
from langchain_databricks import ChatDatabricks
from langchain.agents import Tool, initialize_agent
from pyspark.sql import SparkSession
import yaml
import re

# Initialize Spark and Databricks LLM
spark = SparkSession.builder.getOrCreate()
llm = ChatDatabricks(endpoint="databricks-meta-llama-3-3-70b-instruct", max_tokens=2500)

# Extract only YAML from response
def extract_yaml_from_response(response_text: str) -> str:
    match = re.search(r"```y[a]?ml(.*?)```", response_text, re.DOTALL)
    return match.group(1).strip() if match else response_text.strip()

# Assess summary using LLM
def generate_assessment_summary(script_content: str, script_name: str) -> str:
    prompt = """
            You are an expert in Informatica to Databricks migration. Analyze the given Informatica XML script and summarize it in JSON format.

            1. Count the number of FOLDER nodes in the XML.

            2. For each FOLDER node:
            - If it contains only SOURCE or TARGET nodes, classify it as a "Metadata or Reusability Folder".
            - If it contains any MAPPING nodes, classify it as an "ETL Folder".

            3. For every MAPPING node:
            - Do not generate code. Only generate assessment in the form of a JSON string and store it in a field called `assessment_summary`.

            For each TRANSFORMATION:
            - Extract the `TRANSFORMATION NAME`.
            - Count:
                - Number of input columns (`PORTTYPE="INPUT"` or `"INPUT/OUTPUT"`)
                - Number of output columns (`PORTTYPE="OUTPUT"` or `"INPUT/OUTPUT"`)
                - Number of INSTANCE elements that refer to this transformation
                - Number of CONNECTOR elements involving this transformation
            - **Extract transformation logic** by compiling all non-empty `EXPRESSION` values from `TRANSFORMFIELD` elements. If multiple expressions are present, return them as a list or concatenated string.


            4. Construct a JSON summary with the following json structure:

            {{
            "script_name": "<mapping_name>.xml",
            "folder_name": "<folder_name>",
            "folder_type": "<ETL Folder | Metadata or Reusability Folder>",
            "mappings_info": [
                {{
                "mapping_name": "<mapping_name>",
                "transformation_name": "<transformation_name>",
                "number_of_input_columns": <int>,
                "number_of_output_columns": <int>,
                "number_of_instances": <int>,
                "number_of_connectors": <int>,
                "transformation_logic": "<expression or 'N/A'>"
                }}
            ]
            }}

            5. Return only the JSON string as output. Do not include explanation, commentary, or code.

            Script Content:
            {}
            """.format(script_content)
                
    response = llm.invoke(prompt)
    return response.content.strip()


# Write to Unity Catalog with MERGE based on script_name and type
def write_assessment_to_table(script_name: str, parsed_result: dict, script_type: str, summary: str):
    clean_result = {
        k.lower(): v for k, v in parsed_result.items()
        if k.lower() in ['script_name', 'complexity', 'estimated_effort_days', 'reason']
    }

    df = spark.createDataFrame([{
        "script_name": script_name,
        "complexity": clean_result.get("complexity", "Unknown"),
        "efforts": float(clean_result.get("estimated_effort_days", 0)),
        "reason": clean_result.get("reason", ""),
        "type": script_type,
        "assessment_summary": summary
    }])

    df.createOrReplaceTempView("temp_assessment_result")

    merge_sql = """
    MERGE INTO flash.assessmentout.teradata_to_databricks AS target
    USING temp_assessment_result AS source
    ON target.script_name = source.script_name AND target.type = source.type
    WHEN MATCHED THEN UPDATE SET
        target.complexity = source.complexity,
        target.efforts = source.efforts,
        target.reason = source.reason,
        target.assessment_summary = source.assessment_summary
    WHEN NOT MATCHED THEN INSERT (
        script_name, complexity, efforts, reason, type, assessment_summary
    ) VALUES (
        source.script_name, source.complexity, source.efforts, source.reason, source.type, source.assessment_summary
    )
    """

    print(f"✅ Merging script: {script_name} | type: {script_type}")
    spark.sql(merge_sql)

# Assess script using LLM
def assess_script_with_llm(script_content: str, script_name: str) -> str:
    prompt = f"""
You are an expert in Teradata to Databricks migration. Assess the following script.

Script Name: {script_name}
Script:
{script_content}

Return the result in standard YAML format:
script_name: {script_name}
complexity: <Low|Medium|High>
reason: "<reasoning>"
estimated_effort_days: <number>
"""
    response = llm.invoke(prompt)
    return response.content.strip()

# Generate summary in YAML
def generate_yaml_summary_from_assessments(all_assessments: list) -> str:
    total = len(all_assessments)
    complexity_count = {"Low": 0, "Medium": 0, "High": 0}
    total_effort = 0
    for item in all_assessments:
        comp = item.get("complexity", "Unknown")
        complexity_count[comp] = complexity_count.get(comp, 0) + 1
        total_effort += float(item.get("estimated_effort_days", 0))

    summary = {
        "total_scripts": total,
        "count_by_complexity": complexity_count,
        "total_estimated_effort_days": total_effort,
        "key_observations": "This is a consolidated summary based only on the individual script assessments."
    }
    return yaml.dump(summary, sort_keys=False)

# Run assessment and store output
def assess_scripts_from_table_and_summarize(table: str, script_type_filter: str) -> str:
    try:
        df = spark.table(table).filter(f"type = '{script_type_filter}'")
        rows = df.collect()
        if not rows:
            return f"No scripts of type '{script_type_filter}' found."

        all_assessments = []
        for row in rows:
            script_id = row["id"]
            script = row["script"]
            script_name = f"{script_type_filter}_script_{script_id}.txt"
            print(f"📝 Assessing: {script_name}")
            raw_result = assess_script_with_llm(script, script_name)
            summary = generate_assessment_summary(script, script_name)

            try:
                clean_yaml = extract_yaml_from_response(raw_result)
                parsed_result = yaml.safe_load(clean_yaml)
                print("🧪 Parsed Keys:", list(parsed_result.keys()))
                all_assessments.append(parsed_result)
                write_assessment_to_table(script_name, parsed_result, script_type_filter, summary)
            except yaml.YAMLError:
                print(f"⚠️ Failed to parse YAML for script {script_name}. Cleaned YAML:\n{clean_yaml}")

        if not all_assessments:
            return "No valid assessments could be parsed."

        return generate_yaml_summary_from_assessments(all_assessments)

    except Exception as e:
        return f"❌ Error during assessment: {str(e)}"

# Tool: BTEQ assessment
def assess_bteq_summary_tool(_):
    return assess_scripts_from_table_and_summarize("flash.assessment.teradata_to_databricks", "bteq")

# Tool: Informatica XML assessment
def assess_informatica_summary_tool(_):
    return assess_scripts_from_table_and_summarize("flash.assessment.teradata_to_databricks", "xml")

# Register LangChain tools
tools = [
    Tool(name="assess_bteq_summary", func=assess_bteq_summary_tool, description="Summarize BTEQ script complexity from Unity Catalog."),
    Tool(name="assess_informatica_summary", func=assess_informatica_summary_tool, description="Summarize Informatica job complexity from Unity Catalog.")
]

# Initialize agent
agent = initialize_agent(tools=tools, llm=llm, agent="zero-shot-react-description", verbose=True)

# Entry point
def test_agent():
    print("\n🔍 BTEQ YAML Summary:")
    ##print(agent.run("Summarize complexity of all BTEQ scripts"))

    # Uncomment to test Informatica
    print("\n🔍 Informatica YAML Summary:")
    print(agent.run("Summarize complexity of all Informatica XML jobs"))

if __name__ == "__main__":
    test_agent()
````
