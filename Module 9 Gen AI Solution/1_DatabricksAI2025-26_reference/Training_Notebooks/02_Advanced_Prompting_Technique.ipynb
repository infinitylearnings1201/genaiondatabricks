```python
# Step 2: Import Libraries
from langchain.chat_models import ChatDatabricks
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

# Step 3: Initialize ChatDatabricks LLM
llm = ChatDatabricks(
    endpoint="databricks-meta-llama-3-3-70b-instruct",  # Replace with your actual endpoint
    max_tokens=2048
)

# 💡 Concept: What is Advanced Prompt Engineering?
# Moving from basic prompts to designing prompts that guide the model's internal reasoning, control output format, and simulate expert behavior.

# ---
# 📖 Exercise 1: Chain-of-Thought (CoT) Prompting
cot_prompt = PromptTemplate(
    input_variables=["question"],
    template="""
    You are a smart AI assistant.
    Solve the following question step-by-step:

    Question: {question}

    Let's think step-by-step.
    """
)

cot_chain = LLMChain(llm=llm, prompt=cot_prompt)
question = "If there are 12 apples and you divide them equally among 4 friends, how many apples does each friend get?"
print("--- Chain-of-Thought Reasoning Example ---")
print(cot_chain.run(question=question))

# ---
# 📖 Exercise 2: Role-Based Prompting
role_prompt = PromptTemplate(
    input_variables=["query"],
    template="""
    Assume the role of a university professor explaining to undergrad students.

    Query: {query}

    Answer in simple, engaging, and clear terms with an example.
    """
)

role_chain = LLMChain(llm=llm, prompt=role_prompt)
query = "Why do we use embeddings in machine learning?"
print("--- Role-Based Response Example ---")
print(role_chain.run(query=query))

# ---
# 📖 Exercise 3: Structured Output Prompting
structured_prompt = PromptTemplate(
    input_variables=["concept"],
    template="""
    You are a documentation assistant.
    Provide the following structured explanation for the concept: {concept}

    Respond in this JSON format:
    {{
        "definition": "",
        "key_points": ["", ""],
        "example": ""
    }}
    """
)

structured_chain = LLMChain(llm=llm, prompt=structured_prompt)
concept = "Transfer Learning in AI"
print("--- Structured Output JSON Example ---")
print(structured_chain.run(concept=concept))

# ---
# 📖 Exercise 4: Guardrails and Constraints Prompting
guardrails_prompt = PromptTemplate(
    input_variables=["instruction"],
    template="""
    Follow these strict instructions:

    Instruction: {instruction}

    Constraints:
    - Keep the answer within 3 sentences
    - Avoid technical jargon
    - Do not mention deep learning frameworks (TensorFlow, PyTorch, etc.)
    """
)

guardrails_chain = LLMChain(llm=llm, prompt=guardrails_prompt)
instruction = "Explain what a neural network is."
print("--- Guardrails and Constraint Handling Example ---")
print(guardrails_chain.run(instruction=instruction))

# ✅ Summary for Topic 2: Advanced Prompting Techniques
# In this module, you learned:
# 1. How to design chain-of-thought prompts
# 2. How to simulate roles for different audiences
# 3. How to enforce structured output like JSON
# 4. How to add constraints and guardrails to ensure safe and clear responses

# 📌 Homework:
# - Try creating a few prompts where you combine role + CoT.
# - Try making a JSON structured answer for a real-world business case (like "Explain Customer Lifetime Value").

# Next Topic: Introduction to Retrieval-Augmented Generation (RAG) 🚀
```
