Step 2: Import Libraries

```python
from langchain.chat_models import ChatDatabricks
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

# Step 3: Initialize ChatDatabricks LLM
# NOTE: Replace with your actual serving endpoint
llm = ChatDatabricks(
    endpoint="databricks-meta-llama-3-3-70b-instruct",
    max_tokens=1024
)
```

Concept: What is Prompt Engineering?

Prompt Engineering is the art and science of crafting instructions that guide an LLM to generate useful outputs.

It's like asking the right question to get the most helpful answer.

Step 4: Create a Basic Prompt Template


```python
prompt = PromptTemplate(
    input_variables=["topic"],
    template="""
    You are an expert technical writer.
    Explain the topic: {topic}
    Keep it short, beginner-friendly, and include an analogy.
    """
)

chain = LLMChain(llm=llm, prompt=prompt)
```

```python
# Step 5: Try with Simple Topics
topics = ["Large Language Model", "Overfitting in Machine Learning", "Prompt Engineering"]

for topic in topics:
    print(f"--- Topic: {topic} ---")
    response = chain.run(topic=topic)
    print(response)
    print("\n")
```

```python
# Step 6: Role-Based Prompting
role_prompt = PromptTemplate(
    input_variables=["query"],
    template="""
    You are a senior AI architect mentoring a junior developer.
    Question: {query}
    Provide an answer with a practical example and clear explanation.
    """
)

role_chain = LLMChain(llm=llm, prompt=role_prompt)
query = "How can prompt engineering help improve chatbot responses?"
print("--- Role-based Prompt ---")
print(role_chain.run(query=query))
```

```python
# Step 7: Structured Output Prompting
structured_prompt = PromptTemplate(
    input_variables=["task"],
    template="""
    You are a data assistant. Complete the task and respond in JSON format.

    Task: {task}

    Format:
    {{
        "summary": "...",
        "key_terms": ["...", "..."],
        "difficulty_level": "easy/medium/hard"
    }}
    """
)

structured_chain = LLMChain(llm=llm, prompt=structured_prompt)
task = "Explain the difference between supervised and unsupervised learning."
print("--- Structured Output ---")
print(structured_chain.run(task=task))
```

```python
# Step 8: Chain of Thought Prompting
cot_prompt = PromptTemplate(
    input_variables=["problem"],
    template="""
    You are an AI reasoning assistant.
    Solve the problem step by step and explain your thought process:

    Problem: {problem}

    Let's think step by step.
    """
)

cot_chain = LLMChain(llm=llm, prompt=cot_prompt)
problem = "If a train travels at 60 km/h for 3 hours, how far does it go?"
print("--- Chain of Thought ---")
print(cot_chain.run(problem=problem))
```

```python
# Step 9: Prompt with Constraints and Instruction
constraint_prompt = PromptTemplate(
    input_variables=["instruction"],
    template="""
    Follow the instruction strictly:

    Instruction: {instruction}

    Constraints:
    - Max 3 sentences
    - Avoid technical jargon
    - Use simple language
    """
)

constraint_chain = LLMChain(llm=llm, prompt=constraint_prompt)
instruction = "Describe how neural networks work"
print("--- Prompt with Constraints ---")
print(constraint_chain.run(instruction=instruction))
```
