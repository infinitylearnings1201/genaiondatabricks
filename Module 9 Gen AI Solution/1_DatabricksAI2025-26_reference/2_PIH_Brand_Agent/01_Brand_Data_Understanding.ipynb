 Top Performing KPIs by Average Lift

```python
%sql
SELECT 
  kpi_metric_name, 
  ROUND(AVG(lift_pct), 2) AS avg_lift
FROM solutions_catalog.pih_schema.brand_data
GROUP BY kpi_metric_name
ORDER BY avg_lift DESC
LIMIT 10;
```

Distribution of Studies by Advertiser

```python
%sql
SELECT 
  advertiser_name,
  COUNT(DISTINCT study_id) AS total_studies
FROM solutions_catalog.pih_schema.brand_data
GROUP BY advertiser_name
ORDER BY total_studies DESC;
```

 Lift Trend Over Time by Brand

```python
%sql
SELECT 
  brand_name,
  time_slice_start,
  ROUND(AVG(lift_pct), 2) AS avg_lift
FROM solutions_catalog.pih_schema.brand_data
GROUP BY brand_name, time_slice_start
ORDER BY brand_name, time_slice_start;
```

Compare Exposed vs Control % by Mapping

```python
%sql
SELECT 
  mapping,
  ROUND(AVG(exposed_pct), 2) AS avg_exposed,
  ROUND(AVG(control_pct), 2) AS avg_control
FROM solutions_catalog.pih_schema.brand_data
GROUP BY mapping
ORDER BY avg_exposed DESC;
```

Most Used KPI Metrics by Frequency

```python
%sql

SELECT metric_display_name, COUNT(*) AS usage_count
FROM solutions_catalog.pih_schema.brand_data
GROUP BY metric_display_name
ORDER BY usage_count DESC;
```
