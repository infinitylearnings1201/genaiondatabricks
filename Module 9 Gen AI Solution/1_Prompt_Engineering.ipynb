### What are prompts?
Prompts are the inputs or instructions we give to a Large Language Model (LLM).

They act like the conversation starters or commands that shape the model’s behavior and reasoning.
If an LLM is like a skilled assistant, the prompt is the brief that explains the task — the clearer the brief, the smarter the result.

### 1. Introduction: What Is Prompt Engineering?

Prompt engineering is the art and science of communicating instructions to an LLM clearly and effectively.


**A prompt controls:**

- Task: what the LLM should do
- Persona: how it should behave
- Format: how the answer should look
- Variables: dynamic inputs
- Constraints: rules the model must follow
- Better prompts → better, more consistent outputs.

##3. Import Required Libraries

```python
from langchain_community.chat_models import ChatDatabricks
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
```

### 4. Initialize the Databricks LLM

```python
llm = ChatDatabricks(
    endpoint="databricks-meta-llama-3-3-70b-instruct",
    max_tokens=256
)
```

### 5. PromptTemplate — Simple Text Prompts

```python
template = PromptTemplate.from_template(
    "Explain {topic} in simple terms."
)

prompt = template.format(topic="Lakehouse")
print(prompt)
```

### 6. Prompt Template + LLM (Using LCEL)

Prompt Template + LLM using LCEL means you define a structured prompt (with variables, format, and instructions) and then pass it directly to an LLM through the LangChain Expression Language (LCEL) pipeline.

Ex. Prompt → Model → Output Parser

In a clean, composable way so the LLM always receives a well-formed, consistent instruction every time.

```python
llm = ChatDatabricks(
    endpoint="databricks-meta-llama-3-3-70b-instruct",
    max_tokens=256
)
```

### 7. ChatPromptTemplate — System + User Prompts

```python
chat_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are an expert data engineering instructor."),
    ("human", "Explain {topic} using an analogy.")
])

chain = chat_prompt | llm
print(chain.invoke({"topic": "Delta Lake"}))
```

### 8. Multiple Variables

```python
chat_prompt = ChatPromptTemplate.from_template(
    """
    You are a helpful tutor.

    Topic: {topic}
    Difficulty: {level}

    Provide an easy explanation and one example.
    """
)

chain = chat_prompt | llm
print(chain.invoke({"topic": "Apache Spark", "level": "beginner"}))
```

### 9. Best Practices in Prompt Engineering
**Use system prompts**
Define role & behavior.

**Be explicit**
“Explain in 3 bullet points” > “Explain.”

**Provide structure**
JSON, sections, bullet points.

**Use examples**
Especially for formatting control.

**Keep it short**
Long prompts confuse LLMs more than they help.

### 10. Few-Shot Prompting (Examples)

```python
few_shot = ChatPromptTemplate.from_messages([
    ("system", "Improve poor explanations."),
    ("human", "Bad answer: {bad}\nRewrite it clearly."),
    ("assistant", "Clear version: {example}")
])

chain = few_shot | llm

print(chain.invoke({
    "bad": "Delta Lake is something with tables",
    "example": "Delta Lake is a storage layer that brings ACID reliability to data lakes."
}))
```
