### What are prompts?
Prompts are the inputs or instructions we give to a Large Language Model (LLM).

### What Is Prompt Engineering?
Prompt engineering is the art and science of communicating instructions to an LLM clearly and effectively.

### Topic 2: Advanced Prompt Engineering with LangChain on Databricks

Moving from simple instructions → structured, reliable, multi-step reasoning with LLMs

### 1. Structured Outputs: Teaching the LLM to Follow a Format

LLMs do not naturally follow structure — they need guidance.
If your application expects JSON or a fixed layout, you must explicitly teach the model to follow it.

A good way to begin is to wrap your prompt inside a formatting frame.

```python
%pip install --upgrade langchain langchain-community
dbutils.library.restartPython()
```

```python
from langchain_community.chat_models import ChatDatabricks
from langchain_core.prompts import ChatPromptTemplate

llm = ChatDatabricks(endpoint="databricks-meta-llama-3-3-70b-instruct")

prompt = ChatPromptTemplate.from_template("""
You are an AI that always writes structured JSON.

Explain the topic: {topic}

Return output ONLY in this JSON format:
{{
  "summary": "...",
  "example": "...",
  "keywords": ["...", "...", "..."]
}}
""")

response = (prompt | llm).invoke({"topic": "Delta Lake"})
print(response)
```

Students learn two things here:

- LLMs follow structure when you define it clearly.
- JSON prompts form the backbone of many agentic systems.

This is the foundation for building LLM applications that produce machine-readable outputs, not just text.

### 2. Multi-Step Reasoning: Letting the Model Think in Stages

Sometimes, the answer improves dramatically if the model reasons step by step instead of responding immediately.

You can prompt the model to think through the problem first, then produce the final answer.

```python
prompt = ChatPromptTemplate.from_template("""
First, think step-by-step about the concept: {topic}
Then, after reasoning, provide a final answer in 3 sentences.
""")

print(prompt)
```

```python
response = (prompt | llm).invoke({"topic": "Delta Lake"})
print(response)
```

Encouraging “thinking” improves accuracy, especially for:

- explanations
- transformations
- SQL translation
- code generation
- any domain involving logic

Students start understanding how reasoning bias works inside LLMs.

### 3. Role + Persona Layering: Designing the AI’s Identity

Advanced prompts often assign the model a specific identity to guide tone, depth, and reasoning style.

```python
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a senior data engineer who explains concepts with precision."),
    ("human", "Explain {topic} to a new graduate who knows basic Python but has never used Spark.")
])

response = (prompt | llm).invoke({"topic": "Delta Lake"})
print(response)
```

In systems where clarity is critical (Genie, SQL assistants, multi-agent pipelines), defining persona dramatically improves consistency.

This technique prepares students for agentic work later.

### 4. Prompt Chaining with LCEL: Breaking a Task Into Small Reliable Steps

Large prompts are rarely the solution.
Instead, break a big problem into small modular steps:

- summarize
- extract keywords
- create insights
- write final response

Each step becomes a node in a pipeline.

```python
from langchain_core.prompts import PromptTemplate

step1 = PromptTemplate.from_template("Summarize this text: {text}")
step2 = PromptTemplate.from_template("Extract 5 keywords from this summary: {summary}")
step3 = PromptTemplate.from_template("Create a short insight using: {keywords}")

chain = step1 | llm | step2 | llm | step3 | llm

response = chain.invoke({"text": "India is a diverse country in South Asia known for its rich history, culture, and 'unity in diversity'. It is the world's largest democracy and the seventh-largest country by landmass, bordered by the Himalayas in the north and oceans to the south. Its landscape ranges from mountains to tropical forests, and its population speaks numerous languages and follows various religions, often celebrating different festivals together."})
print(response)
```

This modular thinking is the foundation of LangGraph and multi-agent systems.

Students begin to see LLMs as building blocks, not monolithic black boxes.

### 5. Error Reduction with “Explicit Constraints”

LLMs hallucinate when your instructions are vague. A constraint is a rule that removes ambiguity.

```python
prompt = ChatPromptTemplate.from_template("""
Explain {topic} in exactly:
- 2 bullet points
- No more than 20 words in each bullet
- No analogies
""")

response = (prompt | llm).invoke({"topic": "Delta Lake"})
print(response)

```

Once students learn constraints, they can build assistants that behave predictably.

### 6. Teaching the LLM to Say “I Don’t Know”

This is a critical skill in real-world apps.

A single sentence can reduce hallucinations drastically:

```python
("system", "If you are not sure or lack information, reply 'I don't know' instead of guessing.")
```

For enterprise agents, this is non-negotiable.
